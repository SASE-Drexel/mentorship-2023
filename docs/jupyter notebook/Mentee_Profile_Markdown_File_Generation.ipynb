```python
import pandas as pd
pd.set_option("display.max_columns", 999)
```

```python
mentees = pd.read_csv('mentee-signup.csv')
mentees.head()
```

```python
mentees = mentees.rename(columns={'Briefly Describe yourself (~100 words or less)': "About",
       'Briefly describe your career goals (~50 words or less)': "Career",
       'Briefly describe your interests (~50 words or less)': "Interests",
       'What are you looking for in a mentor? (This could be professional, academic, personal, or anything else, we just want a general idea. ~100 words or less)': "Looking for",})

mentees = mentees.fillna("None")
```

```python
for i in range(0, len(mentees)):
    full_name = mentees['Full Name'][i]
    date = '2023-10-20'
    first_name = mentees['Full Name'][i].split()[0]
    last_name = mentees['Full Name'][i].split()[1]
    year = mentees['Year in Drexel'][i]
    major = mentees['Major'][i]
    about = mentees['About'][i]
    career = mentees['Career'][i]
    interests = mentees['Interests'][i]
    looking_for = mentees['Looking for'][i]
    
    markdown =  '''---
layout: post
title: {} 
date: {}
image: assets/images/{}_{} - {} {}.jpg
about: {}
year: {}
major: {}

---

### About

{}

### Career Goals

{}

### Interests

{}

### Looking For

{}

<div class="text-center my-5">
    <img src="https://sase-drexel.github.io/mentorship-2023/assets/images/{}_{} - {} {}.jpg" alt="{}" class="rounded post-img" />
</div>
'''.format(full_name, date, first_name, last_name, first_name, last_name, about, year, major, about, career, interests, looking_for, first_name, last_name, first_name, last_name, full_name)
    
    mentee_profile = open("{}-{}-{}.md".format(date, first_name, last_name), "w", encoding='utf-8')
    new_profile = mentee_profile.write(markdown)
    mentee_profile.close()
```

```python
mentors = pd.read_csv('mentor-signup.csv')
mentors.head()
```

```python
mentors = mentors.rename(columns={'Briefly describe yourself (<50 words)': "About",
        'Why do you want to be a SASE Mentor? (<100 words)': "Why do I want to be a SASE mentor?",
        'How do you plan to mentor your SASE Mentee? (<100 words)': 'What to expect',
       'What are your career goals? (< 50 words)': "Career",
       'If applicable, list out previous co-ops': "Co-ops",
       'What are some of your interests? (< 50 words)': "Interests",
       'How do you envision your relationship with your mentee?': "Relationship"
        })

mentors = mentors.fillna("None")
```

```python
for i in range(0, len(mentors)):
    full_name = mentors['Full Name'][i]
    date = '2023-10-20'
    first_name = mentors['Full Name'][i].split()[0]
    last_name = mentors['Full Name'][i].split()[1]
    year = mentors['Drexel Year'][i]
    major = mentors['Major'][i]
    about = mentors['About'][i]
    why = mentors['Why do I want to be a SASE mentor?'][i]
    career = mentors['Career'][i]
    co_ops = mentors['Co-ops'][i]
    interests = mentors['Interests'][i]
    what_to_expect = mentors['What to expect'][i]
    relationship = mentors['Relationship'][i]

    markdown =  '''---
layout: post
title: {} 
date: {}
image: assets/images/{}_{} - {} {}.jpg
about: {}
year: {}
major: {}

---

### About

{}

### Why Do I Want To Be a SASE Mentor?

{}

### Career Goals

{}

### Co-ops

{}

### Interests

{}

### What to Expect From Me

{}

### How I Envision My Relationship With My Mentee(s) 

{}

<div class="text-center my-5">
    <img src="https://sase-drexel.github.io/mentorship-2023/assets/images/{}_{} - {} {}.jpg" alt="{}" class="rounded post-img" />
</div>
'''.format(full_name, date, first_name, last_name, first_name, last_name, about, year, major, about, why, career, co_ops, interests, what_to_expect, relationship, first_name, last_name, first_name, last_name, full_name)
    
    mentor_profile = open("{}-{}-{}.md".format(date, first_name, last_name), "w", encoding='utf-8')
    new_profile = mentor_profile.write(markdown)
    mentor_profile.close()
```

```python
mentee_emails = list(mentees['Drexel Email'].values)
mentor_emails = list(mentors['Drexel Email'].values)

print('MENTEE EMAILS')
for mentee in mentee_emails:
    print(mentee, end = ', ')
    
print('\n\nMENTOR EMAILS')
for mentor in mentor_emails:
    print(mentor, end = ', ')
```
